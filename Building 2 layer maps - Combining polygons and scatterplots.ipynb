**Extracting longitude and latitude**

A DataFrame named df has been pre-loaded for you. Complete the code to extract longitude and latitude to new, separate columns.

- Print the first few rows with print() and df.head().
- Use a list comprehension to extract latitude to a new column called lat.
- Use a list comprehension to extract longitude to a new column called lng.
- Print the first few rows with print() and df.head() again to confirm the new columns.

```python
import pandas as pd

data = {
    'StopID': ['4432', '588', '590', '541'],
    'StopName': [
        'MUSIC CITY CENTRAL 5TH - BAY 11',
        'CHARLOTTE AVE & 7TH AVE N WB',
        'CHARLOTTE AVE & 8TH AVE N WB',
        '11TH AVE / N GULCH STATION OUTBOUND',
    ],
    'Location': [
        '(36.16659, -86.781996)',
        '(36.165, -86.78406)',
        '(36.164393, -86.785451)',
        '(36.162249, -86.790464)',
    ],
}

# Buat DataFrame
df = pd.DataFrame(data)

# Ekstrak latitude dan longitude dari kolom "Location"
df['lat'] = df['Location'].str.extract(r'[(](.*?)[,]', expand=False)
df['lon'] = df['Location'].str.extract(r'[,]\s(.*?)[)]', expand=False)

'''
Kita menggunakan metode str.extract dari kolom 'Location' untuk mengekstrak latitude dan longitude. Dalam ekspresi reguler berikut:
- [(] dan [,]\s digunakan untuk mencocokkan tanda kurung buka ( dan tanda koma beserta spasi yang memisahkan latitude dan longitude.
- (.*?) digunakan untuk menangkap nilai latitude dan longitude. 
- (.*?) adalah pola yang mencocokkan sejumlah karakter seminimal mungkin hingga mencapai tanda koma atau tanda kurung tutup, sehingga akan menangkap nilai latitude dan longitude.
Hasil ekstraksi disimpan dalam dua kolom baru, yaitu 'lat' untuk latitude dan 'lon' untuk longitude.
'''


# Konversi latitude dan longitude ke tipe data float
df['lat'] = df['lat'].astype(float)
df['lon'] = df['lon'].astype(float)

print(df)
```

> Nicely done. Now you can pass longitude and latitude more easily to the x and y arguments of a scatterplot.

## Geometry and Shapefile

### Shapefile

shapefile adalah tipe daya khusus yang digunakan untuk menyimpan object geometry berserta dengan atributnya. Tiga object dasar geometry adalah titik, garis dan polygon. Salah satu hal yang menjadi kekurangan dari shapefile adalah pengunaan multi-file dimana ketika menggunakan shapefile tidak hanya `.shp` namun juga diikuti format lainnya dalam satu directory file yang sama. 
- .shp : contains the geometry
- .dbf : holds attributes for each geometry
- .shx : links the attributes to the geometry

> That's right! The geometry field holds a series of latitude/longitude pairs that define the border of a polygon. Printing a geometry field gives all of the latitude/longitude pairs that make up the polygon boundary.
